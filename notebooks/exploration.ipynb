# Trump Company Mention Tracker — Exploration

Load the pipeline output and poke at it. Run `python -m src.pipeline` first.

> Statements are attention/policy signals, **not** investment advice.

```python
import json
from pathlib import Path
import pandas as pd

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
records = [json.loads(l) for l in (ROOT / 'outputs' / 'trump_company_mentions.jsonl').open(encoding='utf-8')]
df = pd.DataFrame(records)
print(len(df), 'records')
df[['date','normalized_company_name','ticker_if_public','sentiment_toward_company','policy_angle','investment_relevance_score','confidence_score']]
```

```python
# Most-mentioned companies
df['normalized_company_name'].value_counts().head(20)
```

```python
# Theme frequency (theme_tags is a list per record)
df.explode('theme_tags')['theme_tags'].value_counts()
```

```python
# High investment-relevance, high-confidence shortlist
hot = df[(df['investment_relevance_score'] >= 4) & (df['confidence_score'] >= 4)]
hot = hot.sort_values(['investment_relevance_score','date'], ascending=False)
hot[['date','normalized_company_name','ticker_if_public','policy_angle','exact_quote']]
```

```python
# Sentiment x policy crosstab
pd.crosstab(df['policy_angle'], df['sentiment_toward_company'])
```
